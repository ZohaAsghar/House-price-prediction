Data cleaning

```python
df=pd.read_csv("content_drive/MyDrive/house_prices.csv")
df=df.drop(['Unnamed: 0'],axis=1)
```

```python
df
```

```python
df.value_counts()
```

```python
df.isnull()
```

```python
df.fillna(2)
```

```python
df['property_type'].value_counts()
```

```python
df['property_type'].value_counts()
df['property_type']=df['property_type'].replace({"House":1,"Flat":2,"Upper Portion":3,"Lower Portion":4,"Penthouse":5,"Room":6,"Farm House":7})
```

```python
df
```

```python
df['location'].value_counts()
```

```python
df['location']=df['location'].replace({"DHA Defence":1,"Bahria Town Karachi":2,"Bahria Town Rawalpindi":3,"Bahria Town":4,"Gulistan-e-Jauhar":5,"Tulspura":6,"Al-Hafeez Garden Road":7,"Airline Avenue":8,"Sitara Supreme City":9,"Gulistan-e-Jinnah Colony":10,"G-10":11,"E-11":12,"G-15":13,"Bani Gala":14})
```

```python
df
```

```python
df["city"].value_counts()
```

```python
df["city"]=df['city'].replace({"Karachi":1,"Lahore":2,"Islamabad":3,"Rawalpindi":4,"Faisalabad":5})
```

```python
df
```

```python
df["purpose"].value_counts()
```

```python
df['purpose']=df['purpose'].replace({"For Sale":1,"For Rent":2})
```

```python
plt.hist(df["city"])
plt.xlabel('City (Encoded)')
plt.ylabel('Frequency')
plt.title('Distribution of Properties by City')
plt.show()
```

```python
# Group 'location' values by 'city' and create a list of Series for boxplot
grouped_data = [df[df['city'] == c]['location'] for c in sorted(df['city'].unique())]

plt.boxplot(grouped_data)

# Set x-axis labels to the actual city encoded values
plt.xticks(ticks=range(1, len(sorted(df['city'].unique())) + 1), labels=sorted(df['city'].unique()))
plt.xlabel('City (Encoded)')
plt.ylabel('Location (Encoded)')
plt.title('Distribution of Location Encoded Values by City')
plt.show()
```

```python
# Group 'location' values by 'city' and create a list of Series for violinplot
grouped_data = [df[df['city'] == c]['location'] for c in sorted(df['city'].unique())]

plt.violinplot(grouped_data)

# Set x-axis labels to the actual city encoded values
plt.xticks(ticks=range(1, len(sorted(df['city'].unique())) + 1), labels=sorted(df['city'].unique()))
plt.xlabel('City (Encoded)')
plt.ylabel('Location (Encoded)')
plt.title('Distribution of Location Encoded Values by City (Violin Plot)')
plt.show()
```
